```python
#Create Rectangle class with attributes length and breadth and methods to find area and perimeter.
#Compare two Rectangle objects by their area. 
class Rectangle:
    def get(self, l, b):
        self.l = l
        self.b = b

    def area(self, l, b):
        return l * b

    def perimeter(self, l, b):
        return 2 * (l + b)


r1 = Rectangle()
r2 = Rectangle()

l1 = int(input("Enter length of Rectangle 1: "))
b1 = int(input("Enter breadth of Rectangle 1: "))
l2 = int(input("Enter length of Rectangle 2: "))
b2 = int(input("Enter breadth of Rectangle 2: "))

p = r1.area(l1, b1)
q = r2.area(l2, b2)

r = r1.perimeter(l1, b1)
s = r2.perimeter(l2, b2)

if p > q:
    print("Rectangle 1 is greater")
elif p < q:
    print("Rectangle 2 is greater")
else:
    print("Rectangles are equal")
print (f"Area of a rectangle 1 ={p},perimeter={r}")
print (f"Area of a rectangle 2 ={q},perimeter={s}")
```

```python
#Create a Bank account with members account number, name, type of account and balance.
#Write constructor and methods to deposit at the bank and withdraw an amount from the bank. 

class BankAccount:
    def __init__(self, a_name, a_no, a_type, a_balance):
        self.acc_name = a_name
        self.acc_no = a_no
        self.acc_type = a_type
        self.acc_balance = a_balance

    def deposit(self, a_deposit):
        print("Initial balance: ", self.acc_balance)
        self.acc_balance += a_deposit
        print(a_deposit, "is deposited")
        print("Current balance: ", self.acc_balance)

    def withdraw(self):
        self.amount = int(input("Withdraw amount: "))
        print("Initial balance: ", self.acc_balance)
        if self.amount > self.acc_balance:
            print("You don't have enough balance to withdraw !!")
            print("Current balance: ", self.acc_balance)
        else:
            print(self.amount, "is withdrawn")
            self.acc_balance -= self.amount
            print("Current balance: ", self.acc_balance)

    def acc_info(self):
        print("Account holder Name: ", self.acc_name)
        print("Account Number: ", self.acc_no)
        print("Account Type: ", self.acc_type)
        print("Account Balance: ", self.acc_balance)


# Create a BankAccount object
name = input("Enter name: ")
no = int(input("Enter account number: "))
acc_type = input("Enter account type: ")
bal = int(input("Enter initial balance: "))
holder = BankAccount(name, no, acc_type, bal)

# Menu-driven interface
while True:
    ch = int(input("1. Deposit 2. Withdraw 3. Account info 4. Exit\nChoose your option: "))
    if ch == 1:
        amount = int(input("Deposit amount: "))
        holder.deposit(amount)
    elif ch == 2:
        holder.withdraw()
    elif ch == 3:
        holder.acc_info()
    elif ch == 4:
        print("Exiting...")
        break
    else:
        print("Invalid option. Please try again.")
```

```python
#Create a class Publisher (name). Derive class Book from Publisher with attributes title and author.
class Publisher:
    def __init__(self, pname):
        self.pubname = pname

    def display1(self):
        print("Book details\nPublisher name:", self.pubname)


class Book(Publisher):
    def __init__(self, pname, bname, author):
        super().__init__(pname)
        self.title = bname
        self.auth = author

    def display2(self):
        print("Title:", self.title)
        print("Author:", self.auth)


class Python(Book):
    def __init__(self, pname, title, au, price, np):
        super().__init__(pname, title, au)
        self.rate = price
        self.nop = np

    def display3(self):
        print("Rate:", self.rate)
        print("No. of pages:", self.nop)


# Input details
p = input("Enter publisher name: ")
b = input("Enter book name: ")
a = input("Enter author name: ")
r = int(input("Enter rate of book: "))
n = int(input("Enter no. of pages: "))

# Create a Python book object and display details
book_obj = Python(p, b, a, r, n)
book_obj.display1()
book_obj.display2()
book_obj.display3()
```

```python
#Create a class Time with private attributes hour, minute and second. 
#Overload ‘+’ operator to find sum of 2 time. 

class Time:
    def __init__(self, hr, mint, sec):
        self.__h = hr
        self.__m = mint
        self.__s = sec

    def __add__(self, other):
        hh = self.__h + other.__h
        mm = self.__m + other.__m
        ss = self.__s + other.__s

        # Handle overflow for seconds
        if ss >= 60:
            m = ss // 60
            ss = ss % 60
            mm += m

        # Handle overflow for minutes
        if mm >= 60:
            h = mm // 60
            mm = mm % 60
            hh += h

        print("Added time..\nHour: ", hh, "Minute: ", mm, "Second: ", ss)


# Input for the first time
h1 = int(input("Enter hour1: "))
m1 = int(input("Enter minute1: "))
s1 = int(input("Enter second1: "))

# Input for the second time
h2 = int(input("Enter hour2: "))
m2 = int(input("Enter minute2: "))
s2 = int(input("Enter second2: "))

# Create Time objects
t1 = Time(h1, m1, s1)
t2 = Time(h2, m2, s2)

# Add the two Time objects
t1 + t2
```
